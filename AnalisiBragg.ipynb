```python
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def normal(t, m, s, N):
    return N/np.sqrt(2*np.pi) * 1/s * np.exp(-0.5*((t-m)/s)**2)

def expon(t, a, b, c):
    return np.exp(b+a*t) + c

def linear(t, a, b):
    return a*t + b
```
