```python
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def normal(t, m, s, N):
    return N/np.sqrt(2*np.pi) * 1/s * np.exp(-0.5*((t-m)/s)**2)

def expon(t, a, b, c):
    return np.exp(b+a*t) + c

def linear(t, a, b):
    return a*t + b
```

```python
def braggKleeman(t, a, b):
    return a/t + b

pr = [650,600,550,500,450]

p, sp = curve_fit(braggKleeman, pr, meansWidth, sigma = sigmasWidth, absolute_sigma = True)

print("Gianca")

residui = np.array(meansWidth - braggKleeman(pr, *p))

fig, ax = plt.subplots(2,1, sharex = True, height_ratios = [1,0.5])
ax[0].errorbar(pr, meansWidth, yerr = sigmasWidth, fmt='o', ms = 1, label = 'Dati')

ax[0].plot(np.arange(450, 650, 1), braggKleeman(np.arange(450, 650, 1), *p), label = "Fit Bragg-Kleeman y = a/x + b")

print("Borto")

ax[0].set_xlabel("p [mbar]")
ax[0].set_ylabel("width [a.u.]")

ax[0].legend()

ax[0].grid(alpha = 0.25)

ax[0].set_title("Altezza picco Bragg vs pressione")

print("Trabba")

# textstr = '\n'.join((
#     r'$a = %.f \pm %.f$ a.u./mbar' % (round(p[0]), round(np.sqrt(sp[0,0]))),
#     r'$b = %.f \pm %.f$ ' % (p[1], np.sqrt(sp[1,1])),
#     r'$\chi^2_{df} = %.2f $' % (np.sum(residui**2/np.array(sigmasWidth)**2))))

print("Bella")

# Posizionamento del box nell'angolo in alto a sinistra
# props = dict(boxstyle='square', facecolor='white', alpha=1)
# ax[0].text(575, 47.5, textstr, transform=ax[0].transAxes,
#            verticalalignment='top', bbox=props)

print("Brenda")

ax[1].errorbar(pr, residui, yerr = sigmasWidth, fmt='o', ms = 1)

ax[1].grid()

print("Niiiiiiiiiiiiiiiick")

plt.show()

print(p, np.sqrt(np.diag(sp)))
```
