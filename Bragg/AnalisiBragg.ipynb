```python
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def normal(t, m, s, N = 1):
    return N/np.sqrt(2*np.pi) * 1/s * np.exp(-0.5*((t-m)/s)**2)

def expon(t, a, b, c):
    return np.exp(b+a*t) + c

def linear(t, a, b):
    return a*t + b
```

```python
integrali = data_p650[" integral"]
fig, ax = plt.subplots()
counts, bins = np.histogram(integrali, bins=800)

wbins = bins[1] - bins[0]

plt.stairs(counts, bins, label='data')

x = []
countAm = 0
for i in range(len(integrali)):
    if integrali[i] > 6500 and integrali[i] < 6900:
        x.append(integrali[i])
        countAm += wbins

mediaAm = np.mean(x)
stdevAm = np.std(x, ddof = 1)

plt.plot(np.arange(6500, 6900), normal(np.arange(6500, 6900), mediaAm, stdevAm, countAm), label = 'picco u='+str(round(mediaAm))+' s='+str(round(stdevAm)))

meansAm.append(mediaAm)
sigmasAm.append(stdevAm)

x = []
countCm = 0
for i in range(len(integrali)):
    if integrali[i] > 6900 and integrali[i] < 7300:
        x.append(integrali[i])
        countCm += wbins
        

mediaCm = np.mean(x)
stdevCm = np.std(x, ddof = 1)

plt.plot(np.arange(6900, 7300), normal(np.arange(6900, 7300), mediaCm, stdevCm, countCm), label = 'picco u='+str(round(mediaCm))+' s='+str(round(stdevCm)))

meansCm.append(mediaCm)
sigmasCm.append(stdevCm)

ax.set_title("Analisi picchi p = 650 mb")

ax.legend()

ax.set_xlim(6100,7400)
ax.set_xlabel("Energy [a.u.]")
ax.set_ylabel("Counts")
plt.plot()

plt.show()
```

```python
integrali = data_p500[" dt"]
fig, ax = plt.subplots()
counts, bins = np.histogram(integrali, bins=200)

plt.stairs(counts, bins, label='data')

p, sp = curve_fit(expon, bins[:-1], counts)
plt.plot(bins[:-1], expon(bins[:-1], *p), label = 'Esponenziale tau= ('+str(round(-p[0],2))+' +/- '+str(round(np.sqrt(sp[0,0]),2))+') s')

ax.fill_between(bins[:-1], expon(bins[:-1], *(p-np.sqrt(np.diag(sp)))), expon(bins[:-1], *(p+np.sqrt(np.diag(sp)))))

ax.set_title("Analisi dt 500 mb")

print(p, np.sqrt(np.diag(sp)))

ax.legend()

ax.set_xlabel("dt")
ax.set_ylabel("Counts")
plt.plot()

plt.show()
```

```python
energy = []
senergy = []
widths = []
swidths = []

integrali = data_p650[" integral"]
width = data_p650[" width"]
fig, ax = plt.subplots()

iAm = []
wAm = []

inf = 6400
sup = 6900

counter = 0
for i in range(len(counts)):
    if integrali[i] > inf and integrali[i] < sup and width[i] > 7 and width[i] < 41:
        iAm.append(integrali[i])
        wAm.append(width[i])
        counter += 1

# plt.plot(np.arange(5400, 6000), normal(np.arange(5400, 6000), *p), label = 'picco Np E= ('+str(int(round(p[0], -1)))+' +/- '+str(int(round(p[1], -1)))+') a.u.')
plt.scatter(iAm, wAm)

ax.set_title("Width vs integral p = 650 mb")

# ax.set_xlim(5000,7400)
ax.set_xlabel("Width [a.u.]")
ax.set_ylabel("Energy [a.u.]")
plt.plot()

plt.show()

fig, ax = plt.subplots()

icounts, ibins = np.histogram(iAm, bins=int(np.sqrt(len(iAm))) + 20)
wcounts, wbins = np.histogram(wAm, bins=int(np.sqrt(len(wAm))) + 20)

imedia = np.mean(iAm)
wmedia = np.mean(wAm)
istd = np.std(iAm, ddof = 1)
wstd = np.std(wAm, ddof = 1)

plt.stairs(icounts, ibins, label='data')

plt.plot(np.arange(inf, sup), normal(np.arange(inf, sup), imedia, istd, counter*(ibins[1]-ibins[0])), label = 'picco Am E= ('+str(int(round(imedia, -1)))+' +/- '+str(int(round(istd, -1)))+') a.u.')

ax.set_title("Integral Am p = 650 mb")

# ax.set_xlim(5000,7400)
ax.set_xlabel("Width [a.u.]")
ax.set_ylabel("Energy [a.u.]")
plt.plot()

energy.append(imedia)
senergy.append(istd)

ax.legend()

plt.show()

fig, ax = plt.subplots()

plt.stairs(wcounts, wbins, label='data')

plt.plot(np.arange(min(wAm), max(wAm),0.05), normal(np.arange(min(wAm), max(wAm),0.05), np.mean(wAm), np.std(wAm, ddof = 1), counter * (wbins[1]-wbins[0])), label = 'picco Am w= ('+str(round(wmedia, 1))+' +/- '+str(round(wstd, 1))+') a.u.')

ax.set_title("Width Am p = 650 mb")

ax.legend()

# ax.set_xlim(5000,7400)
ax.set_xlabel("Width [a.u.]")
ax.set_ylabel("Energy [a.u.]")
plt.plot()

plt.show()

widths.append(wmedia)
swidths.append(wstd)

print(imedia, "+/-", istd)
print(wmedia, "+/-", wstd)
```
